```python
import pandas as pd
```

```python
f = open('1298178.json')
data = json.load(f)
```

```python
print(data.keys())
"""
Meta -> all the meta data
info -> data,event name etc
innings-> ball by ball data
"""
```

```python
title = data['info']['event']['name']
winner = data['info']['outcome']['winner']
toss = data['info']['toss']['winner']
print(f'Title: {title} | Winner: {winner} | Toss-Winner: {toss}')
```

```python
innings_ind = pd.DataFrame(data['innings'][0]['overs'])
```

```python
#  over_count = 0
list_of_overs = []

for over_count in range(0,19):
  one = pd.DataFrame(innings_ind['deliveries'].apply(pd.Series).iloc[0])
  one = one[0].apply(pd.Series)
  two = one['runs'].apply(pd.Series)

  # column Selections for merging to remove duplication column
  one = one[['batter','bowler','non_striker']]
  two = two[['extras','total']]
  over_df = pd.merge(one,two,left_index=True,right_index=True)
  list_of_overs.append(over_df)
```

```python
x = []
x.append(0)
x.append([1,2,3,4])
x.append(over_df)
print(x)
```

```python
india_bat = pd.concat(list_of_overs).fillna(0)
india_bat['ball_count'] = [i for i in range(1,len(india_bat)+1)]
```
