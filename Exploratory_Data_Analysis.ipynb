```python
from collections import Counter

Counts = dict(Counter(df['Reason for absence']))


keys = list(Counts.keys())
values = list(Counts.values())

table = pd.DataFrame()
table['Reason for absence'] = keys
table['count'] = values
table = table.sort_values(by='count',ascending=False).iloc[0:5]
table
```

5 month **Absenteeism**

```python
df = pd.read_csv('Absenteeism_at_work.csv',sep = ';')
```

```python
counts = Counter(df['Month of absence'])
table = pd.DataFrame()
table['month'] = counts.keys()
table['absents'] = counts.values()
table = table.sort_values(by='absents',ascending=False).iloc[0:5]
table 
```

```python
counts = Counter(df['Day of the week'])
table = pd.DataFrame()
table['day'] = counts.keys()
table['absents'] = counts.values()
table = table.sort_values(by='absents',ascending=False).iloc[0:3]
table 
```

```python
days = {
    2:'monday',
    3:'Tuesday',
    4:'Wednesday',
    5:'Thusday',
    6:'Friday',
}
df['Day of the week'] = df['Day of the week'].replace(days)
counts = Counter(df['Day of the week'])

table = pd.DataFrame()
table['day'] = counts.keys()
table['absents'] = counts.values()
table = table.sort_values(by='absents',ascending=False).iloc[0:3]
table 
```
